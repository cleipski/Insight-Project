# Parsing county boundary data

The purpose of this script is to ingest the county boundary data (in long/lat) for counties in the SF Bay area and identifying the min/max values

```python
import os
import glob
import json

%matplotlib inline
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = 15, 6
```

```python
# read counties main file
infile = "README.txt"
infile_path = os.path.join('..','data','county_boundaries',infile)
with open(infile_path,'r') as f:
    lines = f.readlines()
```

```python
# extract county names
counties = []
for line in lines:
    if line[0] == '#':
        pass
    elif len(line.strip()) == 0:
        pass
    else:
        counties.append(line.rstrip())
```

```python
# Read individual files
boundaries = []
final_line_split_all = []
for county in counties:
    tmp_path = os.path.join('..','data','county_boundaries',county,'county_boundary.txt')
    with open(tmp_path,'r') as f:
        lines = f.readlines()  
    final_line = ''
    for line in lines:
        final_line += line
    final_line_split = final_line.split(' ')
    final_line_split_all += final_line_split 
```

```python
# Split final concatenation into a list of long and lat
long_all = []
lat_all = []
for coord in final_line_split_all:
    tmp = coord.split(',')
    long_all.append(tmp[0])
    lat_all.append(tmp[1])
print 'Min/Max value of longitude:', min(long_all), max(long_all)
print 'Min/Max value of latitude:', min(lat_all), max(lat_all)
```

```python
# Give min/max long/lat for each county
out = open(os.path.join('..','data','county_boundaries','min_max.txt'),"w")
out.write('county,long_min,long_max,lat_min,lat_max\n')

for county in counties:
    final_line_split = []
    long_all = []
    lat_all = []
    tmp_path = os.path.join('..','data','county_boundaries',county,'county_boundary.txt')
    with open(tmp_path,'r') as f:
        lines = f.readlines()  
    final_line = ''
    for line in lines:
        final_line += line
    final_line_split = final_line.split(' ')
    for coord in final_line_split:
        tmp = coord.split(',')
        long_all.append(tmp[0])
        lat_all.append(tmp[1])
    print county
    print 'Min/Max value of longitude:', min(long_all), max(long_all)
    print 'Min/Max value of latitude:', min(lat_all), max(lat_all)
    out.write(county+',{},{},{},{}\n'.format(max(long_all),min(long_all),min(lat_all),max(lat_all)))
out.close()
```

```python
# investigate uniqueness of photo_ids
# Get a list of all 'photo_id's
image_id_all = []
counties = ['Marin']
for county in counties:
    tmp_dir = os.path.join('..','data','photo_search',county)
    json_list = glob.glob(os.path.join(tmp_dir,'*json'))
    for tmp_json in json_list:
        # open json file
        with open(tmp_json, 'r') as f:
            data = json.load(f)
        # get image list
        image_list = data['photos']['photo']
        # get all 'image_id's in list
        for image in image_list:
            image_id_all.append((image['id'],county))

```
